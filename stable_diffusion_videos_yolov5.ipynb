```python
from stable_diffusion_videos import walk
```

```python
choose = ['dog',
          'cat',
          'chicken',
          'dinosaur',
          'ford f150',
          'tesla cybertruck',
        
          ]
seed = 57
gen_prompt_sequence = {
}
for i, ob in enumerate(choose):
    template = f"""A showroom displaying a {ob}, environment art,
     landscape art, intricate, hyper detailed, artstation, concept art, wide angle
    , lossless three dimensional ray tracing"""
    gen_prompt_sequence[template] = seed
```

```python
outdir ="/content/yolov5/data/"

walk(prompts=list(gen_prompt_sequence.keys()), 
     seeds=list(gen_prompt_sequence.values()),
     make_video=True,
     output_dir = outdir,
     name='images',
     num_steps=1,
     use_lerp_for_text=True,
     disable_tqdm=True,
     resume=False,
     upsample=True
     )
```
